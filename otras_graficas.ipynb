```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.show

plt.plot(x,y, label='x^2')
plt.title('Simple Line Plot')
plt.xlabel('x Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
plt.plot(x, y, label="x vs y") 

plt.xlabel('Nombre del eje X') 
plt.ylabel('Nombre del eje Y') 
plt.title('Titulo de la grafica') 
plt.legend() 

plt.grid(True) 
```
